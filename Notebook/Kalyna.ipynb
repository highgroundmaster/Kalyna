```python
from sage.crypto.sbox import SBox
import numpy as np
```

```python
aes = [] 
with open(f'Txt_files/aes_sbox.txt', 'rt') as s_box:
    for row in s_box:
        aes += [int(num,16) for num in row.split()]
            
aes = SBox(aes)
aes_ddt = aes.difference_distribution_table()
aes_lat = aes.linear_approximation_table()
```

```python
pi = [[] for i in range(4)]
pi_ddt = [[] for i in range(4)]
pi_lat = [[] for i in range(4)]

for i in range(len(pi)):
    processed_s_box = open(f'Txt_files/s{i+1}.txt', 'w')
    processed_s_box.write("XX," + ','.join(f'{num:02X}' for num in range(16))+ "\n")
        
    
    with open(f'Txt_files/pi{i+1}.txt', 'rt') as s_box:
        for index,row in enumerate(s_box, 0):
            pi[i] += [int(num,16) for num in row.split()]
            processed_s_box.write(f'**{index:02X}**,' + ','.join(row.split()) + "\n")
            
    pi[i] = SBox(pi[i])
    pi_ddt[i] = pi[i].difference_distribution_table()
    pi_lat[i] = pi[i].linear_approximation_table()
    
    with open(f'Txt_files/ddt{i+1}.txt','wb') as ddt:
        for row in np.matrix(pi_ddt[i]):
            np.savetxt(ddt, row, fmt='%#02x',delimiter=', ')
            
    with open(f'Txt_files/lat{i+1}.txt','wb') as lat:
        for row in np.matrix(pi_lat[i]):
            np.savetxt(lat, row, fmt='%#02x',delimiter=', ')
```

```python
def count_diff(s_box, diff):
    sum = 0
    for row in s_box:
        sum += list(row).count(diff)
    return int(sum)
```

```python
for num in range(0,256,2):
    sum = count_diff(list(aes_ddt), num)
    if(sum) :
        print(f'{num} - {sum}')
```

```python
for i in range(4):
    print(f"Pi-{i + 1}")
    for num in range(0,256,2):
        sum = count_diff(list(pi_ddt[i]), num)
        if(sum) :
            print(f'{num} - {sum}')
    print()
```

## Custom Functions

```python
def ntp(num,field):
    '''
    Converts Number to Polynomial in the given Galois Field.
    '''
    return field.fetch_int(num)

    
def ptn(poly,field):
    '''
    Converts Polynomial in the given Galois Field to a Number
    '''
    #Gives the Polynomial Coefficients in Vector Form
    V = field.vector_space(map=False)
    
    #Instantiating given polynomial in Vector Form
    v = V(poly)
    
    #Convert the vector into a Bit string and convert to integer
    num = ''.join(map(str,v))[::-1]
    return int(num, 2)
```

## Mix and Inverse Mix Columns

```python
def mix_and_inv_col(field):
    '''
    Returns a Lookup Table containing Products of Elements of Galois Field with - 0x01,0x04,0x05,0x06,0x07,0xAD, 0x95, 0x76, 0xA8, 0x2F, 0x49, 0xD7, 0xCA
    '''
    #Multiply Mix and Inv Column Elements with all possible elements of field
    mix_inv_mix_elems = [0x01,0x04,0x05,0x06,0x07,0x08,0xAD, 0x95, 0x76, 0xA8, 0x2F, 0x49, 0xD7, 0xCA]
    n = field.order()
    
    #Initializing Table to Zero
    # Using * Operator twice would just created same referenced list causing issues
    prod_tbl = [[0]*n for i in range(len(mix_inv_mix_elems))]
    
    elements = [i for i in range(n)]
    
    
    for elem_1 in range(len(mix_inv_mix_elems)):
        for elem_2 in elements:
            prod_tbl[elem_1][elem_2] = ptn(ntp(mix_inv_mix_elems[elem_1],field)*ntp(elem_2,field),field)
            
    return prod_tbl
```

```python
kalyna_mix_inv_mix = mix_and_inv_col(kalyna_field)

with open('Txt_files/Kalyna_mix_inv_mix.txt','wb') as f:
    for row in np.matrix(kalyna_mix_inv_mix):
        np.savetxt(f, row, fmt='(byte) 0x%02X',delimiter=', ',newline = "},\n{")
```
